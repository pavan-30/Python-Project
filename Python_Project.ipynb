1. What is the overall trend in total sales over the time period covered by the dataset?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
sales_data = pd.read_csv('/content/Auto Sales.csv')

# Convert 'ORDERDATE' column to datetime
sales_data['ORDERDATE'] = pd.to_datetime(sales_data['ORDERDATE'])
sales_data.set_index('ORDERDATE', inplace=True)

# Plot total sales over time
plt.figure(figsize=(15, 6))
sales_data['SALES'].resample('M').sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

- This line plot shows the total sales over time, with the x-axis representing dates and the y-axis representing total sales.
- Insights: We can observe the overall trend in sales over time, including any seasonal patterns or trends.


2. Which product lines contribute the most to total sales?

```python
import numpy as np
import matplotlib.pyplot as plt

# Group sales data by product line and calculate total sales
total_sales_per_product_line = sales_data.groupby('PRODUCTLINE')['SALES'].sum().reset_index()

# Sort product lines by total sales
total_sales_per_product_line = total_sales_per_product_line.sort_values(by='SALES', ascending=False)

# Define colors for bars
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'lightcoral']

# Create the bar plot with custom colors
plt.figure(figsize=(10, 6))
plt.barh(total_sales_per_product_line['PRODUCTLINE'], total_sales_per_product_line['SALES'], color=colors)
plt.title('Sales Distribution by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

- This bar plot displays the total sales for each product line, with the x-axis representing total sales and the y-axis representing product lines.
- Insights: We can identify which product lines contribute the most to the overall sales and compare the sales performance of different product lines.

5. How do sales trends vary by store type?

```python
import matplotlib.pyplot as plt

# Group sales data by order date and product line, then calculate total sales
sales_by_date_and_productline = sales_data.groupby(['ORDERDATE', 'PRODUCTLINE'])['SALES'].sum().reset_index()

# Pivot the data to have product lines as columns
pivot_table = sales_by_date_and_productline.pivot_table(index='ORDERDATE', columns='PRODUCTLINE', values='SALES', aggfunc='sum', fill_value=0)

# Create the stacked area plot
plt.figure(figsize=(15, 6))
pivot_table.plot(kind='area', stacked=True)
plt.title('Sales Trend by Store Type')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend(title='Product Line')
plt.show()
```

- This line plot shows the sales trend over time for different store types, with different colors representing different store types.
- Insights: We can compare the sales performance of different store types over time and identify any trends or patterns specific to each store type.
