```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
sales_data = pd.read_csv('/content/Project/Auto Sales.csv')

# Convert 'ORDERDATE' column to datetime
sales_data['ORDERDATE'] = pd.to_datetime(sales_data['ORDERDATE'])
sales_data.set_index('ORDERDATE', inplace=True)

# Plot total sales over time
plt.figure(figsize=(10, 6))
sales_data['SALES'].resample('M').sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Assuming sales_data contains the dataset with 'PRODUCTLINE' and 'SALES' columns
plt.figure(figsize=(10, 6))
sns.barplot(x='SALES', y='PRODUCTLINE', data=sales_data, estimator=sum)
plt.title('Sales Distribution by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sales_data.groupby('CUSTOMERNAME')['SALES'].mean().sort_values().plot(kind='bar')
plt.title('Average Sales per Customer')
plt.xlabel('Customer')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_selling_products = sales_data.groupby('PRODUCTCODE')['SALES'].sum().nlargest(10)
top_selling_products.plot(kind='bar')
plt.title('Top Selling Products')
plt.xlabel('Product Code')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='ORDERDATE', y='SALES', hue='PRODUCTLINE', data=sales_data)
plt.title('Sales Trend by Store Type')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='SALES', y='COUNTRY', data=sales_data, estimator=sum)
plt.title('Sales Distribution by Country')
plt.xlabel('Total Sales')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEALSIZE', y='SALES', data=sales_data)
plt.title('Sales by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sales_data.resample('M')['SALES'].sum().plot(kind='bar')
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='PRODUCTLINE', y='SALES', data=sales_data)
plt.title('Sales Performance by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sales_data.groupby('CUSTOMERNAME')['SALES'].plot(legend=True)
plt.title('Sales Trend by Customer')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
